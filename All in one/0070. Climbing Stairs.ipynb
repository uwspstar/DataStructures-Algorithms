70. Climbing Stairs

https://leetcode.com/problems/climbing-stairs/description/

You are climbing a staircase. It takes n steps to reach the top.

Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

 

Example 1:

Input: n = 2
Output: 2
Explanation: There are two ways to climb to the top.
1. 1 step + 1 step
2. 2 steps
Example 2:

Input: n = 3
Output: 3
Explanation: There are three ways to climb to the top.
1. 1 step + 1 step + 1 step
2. 1 step + 2 steps
3. 2 steps + 1 step

````python
class solution:
    def climb_stairs(self, n:int) -> int:
        if n == 1: # one stair
            return 1
        if n == 2:
            return 2 # two stair
        
        v1, v2 = 1, 1 # first value = first value + second value

        for _ in range(2, n + 1):
            tmp = v1
            v1 = v1 + v2
            v2 = tmp

        return v1
````

以下是代码的逐行解释，包含中文注释：

```python
class Solution:
    def climb_stairs(self, n: int) -> int:
        # 如果只有 1 个台阶，只有一种走法
        if n == 1:
            return 1
        # 如果有 2 个台阶，有两种走法：一步一步，或者一次两步
        if n == 2:
            return 2
        
        # 初始化 v1 和 v2，分别表示当前和前一个状态
        v1, v2 = 1, 1 # v1 是 f(1)，v2 是 f(0)，即初始状态下爬 1 或 0 个台阶的方式
        
        # 从第 2 级台阶开始，到第 n 级台阶结束
        for _ in range(2, n + 1):
            # 临时存储 v1，便于更新 v2
            tmp = v1
            # 计算当前台阶的走法数，它是前两个台阶的走法数之和
            v1 = v1 + v2
            # 将 v2 更新为旧的 v1 值
            v2 = tmp

        # 返回最终爬到第 n 级台阶的走法数
        return v1
```

### 逐行解释：
1. **处理特殊情况**：
   - 如果只有一个台阶，只有一种走法，返回 1。
   - 如果有两个台阶，可以一次一步或一次两步，有两种走法，返回 2。

2. **初始化状态**：
   - `v1` 和 `v2` 分别初始化为 1，代表爬到第 1 级台阶和第 0 级台阶的走法数。初始状态下，从第 0 级台阶开始，走一步到第 1 级台阶的方式只有一种。
   
3. **循环计算**：
   - 从第 2 级台阶开始，直到第 `n` 级台阶，计算每一级的走法数。每一级的走法数等于前一级和前两级的走法数之和，因为可以从前一级台阶迈一步，或者从前两级台阶迈两步到当前台阶。
   
4. **返回结果**：
   - 当循环结束时，`v1` 存储的是到达第 `n` 级台阶的总走法数。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(n)`，我们遍历了从 2 到 `n` 的台阶数，计算每一级的走法数，因此时间复杂度是线性的。
  
- **空间复杂度**：`O(1)`，只使用了常数空间来存储变量 `v1` 和 `v2`，所以空间复杂度是常数。

因此，整体的时间复杂度是 `O(n)`，空间复杂度是 `O(1)`。

````python
class Solution:
    def climb_stairs(self, n: int) -> int:
        one, two = 1, 1

        for i in range(n - 1):
            tmp = one
            one = one + two
            two = tmp
        
        return one
````
