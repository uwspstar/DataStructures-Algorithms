198. House Robber

https://leetcode.com/problems/house-robber/description/

You are a professional robber planning to rob houses along a street. Each house has a certain amount of money stashed, the only constraint stopping you from robbing each of them is that adjacent houses have security systems connected and it will automatically contact the police if two adjacent houses were broken into on the same night.

Given an integer array nums representing the amount of money of each house, return the maximum amount of money you can rob tonight without alerting the police.

 

Example 1:

Input: nums = [1,2,3,1]
Output: 4
Explanation: Rob house 1 (money = 1) and then rob house 3 (money = 3).
Total amount you can rob = 1 + 3 = 4.
Example 2:

Input: nums = [2,7,9,3,1]
Output: 12
Explanation: Rob house 1 (money = 2), rob house 3 (money = 9) and rob house 5 (money = 1).
Total amount you can rob = 2 + 9 + 1 = 12.

````python
class Solution:
    def rob(self, nums: List[int]) -> int:
        rob1, rob2 = 0, 0
        #[rob1, rob2, n, n + 1, ...]
        for n in nums:
            tmp = max(rob1 + n, rob2)
            rob1 = rob2
            rob2 = tmp

        return rob2
````

    def rob(self, nums: List[int]) -> int:
        # rob1 和 rob2 初始化为 0。
        # rob1 表示当前房子前两步的最大抢劫金额。

            # 或者跳过当前房子，保留目前为止的最大抢劫金额（rob2）。

            # 将 rob1 移动到 rob2 的值（因为我们向前移动了一步，rob2 现在变成 rob1）。

### 解释：
1. **初始化**：`rob1` 和 `rob2` 用于存储截至前两座房子和前一座房子的最大抢劫金额。
2. **遍历列表**：对于每一个房子，它计算能够抢到的最大金额，分为两种情况：
   - 抢劫当前房子（将当前房子金额加到前两座房子的最大金额上，即 `rob1 + n`）。
   - 跳过当前房子，保留前一座房子的最大金额（`rob2`）。
3. **更新值**：每经过一个房子，`rob1` 更新为 `rob2` 的值（向前移动一步），`rob2` 更新为计算出的最大抢劫金额（`tmp`）。
4. **最终结果**：遍历所有房子后，`rob2` 存储的是最大抢劫金额，同时不违反不能连续抢劫两座房子的规则。

这是一个解决“打家劫舍”问题的动态规划解法，要求不能抢劫相邻的房子。

以下是代码的逐行解释，包含中文注释：

```python
class Solution:
    def rob(self, nums: List[int]) -> int:
        # 初始化 rob1 和 rob2 分别为 0，表示截至前两家和截至前一家房子的最大抢劫金额
        rob1, rob2 = 0, 0
        
        # [rob1, rob2, n, n + 1, ...] 是一个提示，表示每次选择最大值时的逻辑
        # 遍历每一家房子的金额
        for n in nums:
            # tmp 存储抢劫当前房子和前两家房子金额的最大值，或跳过当前房子
            tmp = max(rob1 + n, rob2)
            # 更新 rob1 为 rob2 的值（相当于移动到下一个房子）
            rob1 = rob2
            # 更新 rob2 为 tmp，即包括当前房子的最大抢劫金额
            rob2 = tmp

        # 返回 rob2，即最大可能的抢劫金额
        return rob2
```

### 逐行解释：
1. **初始化**：
   - `rob1` 和 `rob2` 都初始化为 0，表示前两家和前一家房子的最大抢劫金额。在开始时，我们假设没有任何房子，因此都为 0。
   
2. **遍历 `nums` 列表**：
   - 对于每个房子的金额 `n`，我们计算在这一家房子时可以选择的两种方案：
     1. **抢劫当前房子**，那么总抢劫金额是 `rob1 + n`（前两家房子的最大金额加上当前房子的金额）。
     2. **不抢劫当前房子**，那么总抢劫金额就是 `rob2`（前一家房子的最大金额）。
   
   - `tmp` 存储这两种方案的最大值。

3. **更新状态**：
   - `rob1` 更新为 `rob2`，因为下一步的前两家房子的最大值是当前 `rob2` 的值。
   - `rob2` 更新为 `tmp`，表示包括当前房子的最大抢劫金额。

4. **返回结果**：
   - 遍历完所有房子后，`rob2` 包含了最大可能的抢劫金额。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(n)`，其中 `n` 是 `nums` 列表的长度。我们只遍历一次房子列表，进行常数时间的计算。
  
- **空间复杂度**：`O(1)`，因为只使用了常数空间存储 `rob1`、`rob2` 和 `tmp`。

因此，这个算法的时间复杂度为 `O(n)`，空间复杂度为 `O(1)`。这个问题是典型的动态规划问题，每次选择是否抢劫当前房子，依赖于前面的状态。
