Given an array arr, replace every element in that array with the greatest element among the elements to its right, and replace the last element with -1.

After doing so, return the array.

 

Example 1:

Input: arr = [17,18,5,4,6,1]
Output: [18,6,6,6,1,-1]
Explanation: 
- index 0 --> the greatest element to the right of index 0 is index 1 (18).
- index 1 --> the greatest element to the right of index 1 is index 4 (6).
- index 2 --> the greatest element to the right of index 2 is index 4 (6).
- index 3 --> the greatest element to the right of index 3 is index 4 (6).
- index 4 --> the greatest element to the right of index 4 is index 5 (1).
- index 5 --> there are no elements to the right of index 5, so we put -1.
Example 2:

Input: arr = [400]
Output: [-1]
Explanation: There are no elements to the right of index 0.



````python
class Solution:
    def replaceElements(self, arr: List[int]) -> List[int]:
        max_right =  -1
        for i in range(len(arr) -1, -1, -1):
            new_max = max (max_right, arr[i])
            arr[i] = max_right
            max_right = new_max
        return arr
````

以下是代码的逐行解释，包含中文注释：

```python
class Solution:
    def replaceElements(self, arr: List[int]) -> List[int]:
        # 初始化 max_right 为 -1，用来记录右边的最大值
        max_right = -1
        
        # 从数组的最后一个元素开始遍历，向前遍历
        for i in range(len(arr) - 1, -1, -1):
            # 记录当前元素和右边最大值之间的较大值
            new_max = max(max_right, arr[i])
            
            # 将当前元素替换为右边的最大值
            arr[i] = max_right
            
            # 更新右边的最大值为新的最大值
            max_right = new_max
        
        # 返回修改后的数组
        return arr
```

### 逐行解释：

1. **初始化**：
   - `max_right = -1`：右边的最大值初始化为 `-1`，因为数组最后一个元素的右边没有元素，所以初始化为 `-1`。

2. **遍历数组**：
   - `for i in range(len(arr) - 1, -1, -1)`：从数组的最后一个元素开始，逆序遍历数组（从右向左）。
   
3. **更新当前元素和最大值**：
   - `new_max = max(max_right, arr[i])`：在每个位置，比较当前元素 `arr[i]` 和右边的最大值 `max_right`，更新 `new_max` 为它们之间的较大值。
   
4. **替换当前元素**：
   - `arr[i] = max_right`：将当前元素替换为右边最大值 `max_right`，然后更新 `max_right` 为 `new_max`，以便在下一个元素中使用。

5. **更新最大值**：
   - `max_right = new_max`：将 `max_right` 更新为当前的 `new_max`，以便在接下来处理左边的元素时使用。

6. **返回结果**：
   - `return arr`：返回处理后的数组。

### 代码逻辑：

- 在每个元素的位置上，代码将该元素替换为其右边所有元素中的最大值。
- 因为从右向左遍历，因此可以使用一个变量 `max_right` 来存储当前右边的最大值，避免重复遍历右边的元素。

### 时间和空间复杂度：

- **时间复杂度**：`O(n)`，其中 `n` 是数组的长度。我们只遍历数组一次，所以时间复杂度是线性的。

- **空间复杂度**：`O(1)`，我们只使用了一个额外的变量 `max_right`，因此空间复杂度是常数。

### 总结：
该算法通过从右向左遍历数组，逐个替换当前元素为右边的最大值。由于只需要遍历一次数组，并使用了常数级别的额外空间，时间复杂度为 `O(n)`，空间复杂度为 `O(1)`，是一种高效的解决方案。
