53. Maximum Subarray

https://leetcode.com/problems/maximum-subarray/description/

Given an integer array nums, find the 
subarray
 with the largest sum, and return its sum.

 

Example 1:

Input: nums = [-2,1,-3,4,-1,2,1,-5,4]
Output: 6
Explanation: The subarray [4,-1,2,1] has the largest sum 6.
Example 2:

Input: nums = [1]
Output: 1
Explanation: The subarray [1] has the largest sum 1.
Example 3:

Input: nums = [5,4,-1,7,8]
Output: 23
Explanation: The subarray [5,4,-1,7,8] has the largest sum 23.

````python
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        max_sum, curr_sum = nums[0], 0
        for num in nums:
            if curr_sum < 0:
                curr_sum = 0
            curr_sum += num
            max_sum = max(max_sum, curr_sum) 
        return max_sum
````

以下是代码的逐行解释，包含中文注释：

```python
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        # 初始化 max_sum 为数组中的第一个元素，curr_sum 为 0
        # max_sum 用来存储迄今为止的最大子数组和，curr_sum 用来存储当前子数组的和
        max_sum, curr_sum = nums[0], 0

        # 遍历数组 nums 中的每一个元素
        for num in nums:
            # 如果当前子数组的和 curr_sum 小于 0，就将其重置为 0
            # 因为负数的和会拖累接下来的子数组和，因此可以抛弃它
            if curr_sum < 0:
                curr_sum = 0

            # 累加当前元素到 curr_sum 中
            curr_sum += num

            # 更新 max_sum，比较 max_sum 和当前子数组和 curr_sum 的大小，取较大值
            max_sum = max(max_sum, curr_sum)

        # 返回最大子数组和 max_sum
        return max_sum
```

### 逐行解释：
1. **初始化**：
   - `max_sum` 被初始化为数组的第一个元素，这保证即使所有的数都是负数，结果也不会比数组中的最大元素小。
   - `curr_sum` 初始化为 0，表示当前子数组的和，将用来累加当前子数组的元素。

2. **遍历数组**：
   - 对于每个 `num`：
     - 如果 `curr_sum` 小于 0，则重置为 0，意味着抛弃当前的子数组（因为一个负数的和只会降低后续的总和）。
     - 将当前的 `num` 加入到 `curr_sum`，表示将当前数包含到当前子数组中。
     - 通过 `max()` 函数来比较 `max_sum` 和 `curr_sum`，并更新 `max_sum` 为更大值。

3. **返回结果**：
   - 最终返回 `max_sum`，它表示整个数组中找到的最大子数组和。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(n)`，其中 `n` 是 `nums` 列表的长度。因为我们只遍历 `nums` 一次，所有操作都是在常数时间内完成的。
  
- **空间复杂度**：`O(1)`，我们只使用了几个额外的变量（`max_sum` 和 `curr_sum`），因此空间复杂度是常数。

因此，整体的时间复杂度是 `O(n)`，空间复杂度是 `O(1)`。
