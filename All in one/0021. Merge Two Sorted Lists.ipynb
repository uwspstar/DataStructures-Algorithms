21. Merge Two Sorted Lists

https://leetcode.com/problems/merge-two-sorted-lists/description/

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists in a one-sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

Example:

Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
Example:

Input: list1 = [], list2 = []
Output: []
Example:

Input: list1 = [], list2 = [0]
Output: [0]

````python
from typing import Optional 


class ListNode: # 1
    def __init__(self, val=0, next=None) -> None:
        self.val = val
        self.next =  next

class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2:Optional[ListNode]):
        dummy = tail = ListNode(0)

        while list1 and list2:
            if list1.val < list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 =  list2.next
            
            tail = tail.next
        
        tail.next = list1 or list2

        return dummy.next
````

````python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None) -> None:
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, list1:Optional[ListNode], list2:Optional[ListNode]) -> Optional[ListNode]:
        if not list1:
            return list2
        if not list2:
            return list1
        
        if list1.val < list2.val:
            list1.next = self.mergeTwoLists(list1.next, list2)
            return list1
        else:
            list2.next = self.mergeTwoLists(list1, list2.next)
            return list2
````

```python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None) -> None:
        # 初始化节点，val 是节点的值，next 是指向下一个节点的指针
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # 如果 list1 为空，直接返回 list2，因为没有东西可合并
        if not list1:
            return list2
        # 如果 list2 为空，直接返回 list1，情况同上
        if not list2:
            return list1
        
        # 如果 list1 的值小于 list2 的值
        if list1.val < list2.val:
            # 递归地合并 list1 的下一个节点和 list2，并将结果作为 list1 的 next
            list1.next = self.mergeTwoLists(list1.next, list2)
            # 返回 list1，因为它是当前的较小值节点
            return list1
        else:
            # 递归地合并 list1 和 list2 的下一个节点，并将结果作为 list2 的 next
            list2.next = self.mergeTwoLists(list1, list2.next)
            # 返回 list2，因为它是当前的较小值节点
            return list2
```

### 逐行解释：
1. **`ListNode` 类**：定义链表节点，`val` 是节点的值，`next` 是指向下一个节点的指针。该类用于创建链表的节点。

2. **`mergeTwoLists` 方法**：
   - **边界条件**：首先，处理 `list1` 或 `list2` 为空的情况：
     - 如果 `list1` 为空，说明剩余的只有 `list2`，直接返回 `list2`。
     - 如果 `list2` 为空，说明剩余的只有 `list1`，直接返回 `list1`。
   
   - **递归合并**：
     - 如果 `list1` 的当前值小于 `list2` 的当前值，递归地将 `list1` 的下一个节点与 `list2` 合并，并将结果设为 `list1.next`，然后返回 `list1`。
     - 如果 `list2` 的值小于或等于 `list1` 的值，递归地将 `list1` 与 `list2` 的下一个节点合并，并将结果设为 `list2.next`，然后返回 `list2`。

3. **递归逻辑**：这个方法通过递归比较 `list1` 和 `list2` 的当前节点值，并将较小值的节点保留在结果链表中，然后继续合并剩余部分。最终，递归会到达一个链表为空的情况，开始返回合并后的链表。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(n + m)`，其中 `n` 和 `m` 分别是 `list1` 和 `list2` 的长度。我们在递归中处理每个节点一次，所以时间复杂度是线性的。
  
- **空间复杂度**：`O(n + m)`，递归的调用栈深度最多为两个链表的长度和，因此空间复杂度为 `O(n + m)`。

因此，时间复杂度为 `O(n + m)`，空间复杂度为 `O(n + m)`。
