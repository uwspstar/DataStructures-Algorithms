226. Invert Binary Tree

https://leetcode.com/problems/invert-binary-tree/description/

Given the root of a binary tree, invert the tree, and return its root.

 

Example 1:


Input: root = [4,2,7,1,3,6,9]
Output: [4,7,2,9,6,3,1]
Example 2:


Input: root = [2,1,3]
Output: [2,3,1]
Example 3:

Input: root = []
Output: []
 

Constraints:

The number of nodes in the tree is in the range [0, 100].
-100 <= Node.val <= 100

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:
            return root
        
        left = self.invertTree(root.left)
        right = self. invertTree(root.right)

        root.left, root.right = right, left

        return root
````

以下是代码的逐行解释，包含中文注释：

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        # 如果根节点为空，直接返回（空树或递归终止条件）
        if not root:
            return root
        
        # 递归翻转左子树
        left = self.invertTree(root.left)
        # 递归翻转右子树
        right = self.invertTree(root.right)

        # 将左右子树交换
        root.left, root.right = right, left

        # 返回翻转后的根节点
        return root
```

### 逐行解释：

1. **基准条件**：
   - `if not root`: 如果当前节点为空，说明已到达叶子节点的末端，直接返回 `None`，作为递归的终止条件。

2. **递归调用**：
   - `left = self.invertTree(root.left)`：递归翻转左子树，并将结果赋给 `left`。
   - `right = self.invertTree(root.right)`：递归翻转右子树，并将结果赋给 `right`。

3. **交换子树**：
   - `root.left, root.right = right, left`: 将当前节点的左子树和右子树交换，达到树的翻转效果。

4. **返回结果**：
   - `return root`：返回当前节点（已经完成翻转的节点），继续进行上层递归。

### 时间和空间复杂度：

- **时间复杂度**：`O(n)`，其中 `n` 是树中节点的数量。每个节点被访问一次，因此时间复杂度是线性的。

- **空间复杂度**：`O(h)`，其中 `h` 是树的高度。由于递归栈的深度最多为树的高度，在最坏情况下（树呈现链表结构），空间复杂度为 `O(n)`。

### 解释：
该算法通过递归实现树的翻转。在每个递归调用中，分别翻转当前节点的左右子树，然后交换它们，达到翻转整个二叉树的目的。最终返回翻转后的根节点。

### 总结：
该递归算法以 `O(n)` 的时间复杂度和 `O(h)` 的空间复杂度实现了二叉树的翻转。递归的思路清晰，通过交换左右子树，逐步完成整个树的翻转。
