121. Best Time to Buy and Sell Stock

https://leetcode.com/problems/best-time-to-buy-and-sell-stock/description/

you are given an array prices where prices[i] is the price of a given stock on the ith day.

You want to maximize your profit by choosing a single day to buy one stock and choosing a different day in the future to sell that stock.

Return the maximum profit you can achieve from this transaction. If you cannot achieve any profit, return 0.

 

Example 1:

Input: prices = [7,1,5,3,6,4]
Output: 5
Explanation: Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5.
Note that buying on day 2 and selling on day 1 is not allowed because you must buy before you sell.
Example 2:

Input: prices = [7,6,4,3,1]
Output: 0
Explanation: In this case, no transactions are done and the max profit = 0.




````python
from typing import List


class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        max_profit = 0
        buy = prices[0]

        for sell in prices[1:]:
            if sell > buy:
                max_profit = max(max_profit, sell - buy)
            else:
                buy = sell 

        return max_profit
````

以下是代码的逐行解释，包含中文注释：

```python
from typing import List

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        # 初始化 max_profit 为 0，表示没有利润
        max_profit = 0
        # 初始化 buy 为第一个股票价格，即第一天的买入价格
        buy = prices[0]

        # 从第二天开始遍历价格，尝试卖出
        for sell in prices[1:]:
            # 如果当前卖出价格大于买入价格，计算当前利润并更新最大利润
            if sell > buy:
                max_profit = max(max_profit, sell - buy)
            else:
                # 如果当前卖出价格小于或等于买入价格，更新买入价格为当前价格
                buy = sell

        # 返回最终计算出的最大利润
        return max_profit
```

### 逐行解释：
1. **初始化**：
   - `max_profit` 初始化为 0，表示开始时没有任何利润。
   - `buy` 初始化为第一个价格 `prices[0]`，表示在第一天买入股票。

2. **遍历股票价格**：
   - 从第二天开始，通过遍历 `prices[1:]`，每次计算当日卖出时的利润。
   - 如果当前价格 `sell` 高于 `buy`（即存在利润），计算利润 `sell - buy`，并更新 `max_profit` 为更大值。
   - 如果当前价格 `sell` 小于或等于 `buy`，则更新 `buy` 为当前价格，因为这意味着我们找到了一个更低的买入价格，后续的潜在利润可能会增加。

3. **返回结果**：
   - 遍历结束后，返回 `max_profit`，表示我们能够获取的最大利润。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(n)`，其中 `n` 是 `prices` 列表的长度。我们只需要遍历一次价格数组，因此时间复杂度是线性的。
  
- **空间复杂度**：`O(1)`，我们只使用了常数空间来存储 `max_profit` 和 `buy`，因此空间复杂度是常数。

### 总结：
- 时间复杂度为 `O(n)`，因为我们只遍历了价格列表一次。
- 空间复杂度为 `O(1)`，因为只使用了几个额外的变量。

这个解决方案是通过一次遍历找到最优的买卖时机，从而获得最大利润的贪心算法。
