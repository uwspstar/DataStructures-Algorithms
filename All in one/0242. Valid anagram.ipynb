242. Valid Anagram

https://leetcode.com/problems/valid-anagram/description/

Given two strings s and t, return true if t is an 
anagram
 of s, and false otherwise.

 

Example 1:

Input: s = "anagram", t = "nagaram"

Output: true

Example 2:

Input: s = "rat", t = "car"

Output: false

````python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        cntS, cntT = {}, {}

        for c in s:
            cntS[c] = cntS.get(c, 0) + 1
        
        for c in t:
            cntT[c] = cntT.get(c, 0) + 1
        
        for key in cntS:
            if cntS[key] != cntT.get(key, 0):
                return False
        
        return True       
````

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # 如果两个字符串的长度不相等，它们不可能是字母异位词，直接返回 False
        if len(s) != len(t):
            return False

        # 创建两个字典 cntS 和 cntT，分别用于存储 s 和 t 中每个字符的出现次数
        cntS, cntT = {}, {}

        # 遍历字符串 s，将每个字符的出现次数记录在 cntS 中
        for c in s:
            cntS[c] = cntS.get(c, 0) + 1
        
        # 遍历字符串 t，将每个字符的出现次数记录在 cntT 中
        for c in t:
            cntT[c] = cntT.get(c, 0) + 1
        
        # 遍历 cntS 中的每个键（字符），检查它在 cntT 中的出现次数是否相同
        for key in cntS:
            # 如果 cntS 中的某个字符在 cntT 中的计数不同，返回 False
            if cntS[key] != cntT.get(key, 0):
                return False
        
        # 如果所有字符的出现次数都相同，则返回 True，表示 s 和 t 是字母异位词
        return True
```

### 逐行解释：
1. **长度检查**：
   - 如果 `s` 和 `t` 的长度不同，它们不可能是字母异位词，直接返回 `False`。

2. **字符计数**：
   - 使用字典 `cntS` 和 `cntT` 分别记录两个字符串中每个字符的出现次数。
   - 通过 `get()` 方法获取当前字符的出现次数，如果字符不存在则返回默认值 `0`，然后递增该字符的计数。

3. **比较字符计数**：
   - 遍历 `cntS` 中的每个字符，检查它在 `cntT` 中的出现次数是否相同。
   - 如果发现某个字符在两个字符串中的出现次数不同，立即返回 `False`。

4. **返回结果**：
   - 如果所有字符的出现次数都匹配，返回 `True`，表示 `s` 和 `t` 是字母异位词。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(n)`，其中 `n` 是字符串 `s` 或 `t` 的长度。我们需要遍历两个字符串来构建字符计数字典，然后再遍历字典进行比较。
  
- **空间复杂度**：`O(1)`，因为字母异位词只涉及有限的字符集（通常为 26 个小写字母），因此计数字典的大小是固定的。但严格来说，空间复杂度是与字符集大小成正比的，若字符集无限大，则空间复杂度为 `O(n)`。

### 优化建议：
可以使用一个单一的字典来计数字符，而不是分别为两个字符串构建字典。通过对一个字符串的字符计数递增，另一个字符串的字符计数递减，最后检查字典中是否所有值为 `0` 即可。这样可以减少对字典的遍历。

是有一种更好的方法来检查两个字符串是否为字母异位词。我们可以通过一个单一的计数器来完成，而不需要分别为每个字符串构建两个字典。

### 优化思路：
1. **单一计数器**：我们可以使用一个字典来计数字符。在遍历第一个字符串时，增加对应字符的计数；在遍历第二个字符串时，减少对应字符的计数。
2. **检查计数器**：最后检查字典中的所有计数是否为 0。如果所有计数为 0，说明两个字符串是字母异位词。

### 优化后的代码：

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # 如果两个字符串的长度不相等，直接返回 False
        if len(s) != len(t):
            return False

        # 创建一个字典用于计数
        count = {}

        # 遍历字符串 s，增加每个字符的计数
        for c in s:
            count[c] = count.get(c, 0) + 1

        # 遍历字符串 t，减少每个字符的计数
        for c in t:
            if c not in count:  # 如果字符不在字典中，直接返回 False
                return False
            count[c] -= 1
            if count[c] < 0:  # 如果某个字符的计数小于 0，说明 t 中有多余的字符，返回 False
                return False

        # 如果所有字符的计数都为 0，返回 True
        return True
```

### 优化后的思路：
1. **长度检查**：首先检查两个字符串长度是否相等，如果长度不同，立即返回 `False`。
2. **计数字符**：使用一个字典 `count` 记录字符的出现次数。
   - 在遍历字符串 `s` 时，增加字符的计数。
   - 在遍历字符串 `t` 时，减少字符的计数。
3. **提前退出条件**：
   - 如果遍历 `t` 时发现某个字符不在字典中，或者该字符的计数变为负值，表示 `t` 中有额外的字符，立即返回 `False`。
4. **最终检查**：由于我们在遍历 `t` 时就处理了所有可能的情况，因此不需要进一步检查，只要两个字符串的字符计数相符即可返回 `True`。

### 优化后的时间和空间复杂度：
- **时间复杂度**：`O(n)`，其中 `n` 是字符串的长度。我们仍然需要遍历每个字符串一次。
- **空间复杂度**：`O(1)`，使用的字典最多只会包含 26 个小写字母的键值对，因此空间复杂度是常数（假设字符集是固定的）。

相比于原始方法，这种方式减少了额外的遍历和字典操作，且更加简洁高效。
