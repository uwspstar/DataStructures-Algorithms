203. Remove Linked List Elements

https://leetcode.com/problems/remove-linked-list-elements/description/

Given the head of a linked list and an integer val, remove all the nodes of the linked list that has Node.val == val, and return the new head.

 

Example 1:


Input: head = [1,2,6,3,4,5,6], val = 6
Output: [1,2,3,4,5]
Example 2:

Input: head = [], val = 1
Output: []
Example 3:

Input: head = [7,7,7,7], val = 7
Output: []
 

Constraints:

The number of nodes in the list is in the range [0, 104].
1 <= Node.val <= 50
0 <= val <= 50


````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:

        dummy = ListNode(0, next = head)
        pre, cur = dummy, head

        while cur:
            if cur.val == val:
                pre.next = cur.next
                cur = cur.next
            else: 
                pre = cur
                cur = cur.next
        return dummy.next
            
        
````

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        # 创建一个虚拟头节点，next 指向原链表的头节点
        dummy = ListNode(0, next=head)
        # 初始化两个指针，pre 表示前一个节点，cur 表示当前节点
        pre, cur = dummy, head

        # 开始遍历链表
        while cur:
            # 如果当前节点的值等于要删除的目标值
            if cur.val == val:
                # 前一个节点的 next 指针指向当前节点的下一个节点
                pre.next = cur.next
                # 当前节点移动到下一个节点
                cur = cur.next
            else:
                # 如果当前节点的值不等于目标值，pre 和 cur 都向前移动
                pre = cur
                cur = cur.next
        # 返回虚拟头节点的 next，跳过虚拟头节点
        return dummy.next
```

### 逐行解释：
1. **定义虚拟头节点**：
   - **`dummy = ListNode(0, next=head)`**：创建一个虚拟的头节点 `dummy`，并让它的 `next` 指向原链表的头节点 `head`。这是为了方便处理头节点需要被删除的情况。
   - **`pre, cur = dummy, head`**：初始化两个指针，`pre` 表示前一个节点，`cur` 表示当前节点，分别指向虚拟头节点和链表的实际头节点。

2. **遍历链表**：
   - 使用 **`while cur:`** 循环，遍历整个链表，直到 `cur` 为空。
   - **`if cur.val == val:`**：如果当前节点 `cur` 的值等于目标值 `val`，则进行删除操作。
     - **`pre.next = cur.next`**：将 `pre` 的 `next` 指针指向 `cur` 的下一个节点，从而跳过当前节点。
     - **`cur = cur.next`**：移动当前节点到下一个节点。
   - **`else:`**：如果当前节点不需要删除，则将 `pre` 和 `cur` 同时向前移动一位，继续遍历。

3. **返回结果**：
   - **`return dummy.next`**：返回虚拟头节点的 `next`，即跳过虚拟头节点的实际链表头部，得到处理后的链表。

- **时间复杂度**：`O(n)`，其中 `n` 是链表的长度。我们需要遍历链表中的每个节点一次，因此时间复杂度为线性。
- **空间复杂度**：`O(1)`，我们只使用了常量级的额外空间（两个指针 `pre` 和 `cur`）。

- **虚拟头节点**：通过引入一个虚拟头节点，处理链表删除操作时可以避免特殊处理头节点的情况。
- **链表遍历与删除**：通过两个指针 `pre` 和 `cur` 遍历链表，在发现目标值时跳过相应节点。

### 小结：

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        # 如果链表为空，直接返回
        if not head:
            return head
        
        # 递归处理链表的下一个节点
        head.next = self.removeElements(head.next, val)

        # 如果当前节点的值等于要删除的值，跳过当前节点
        # 否则保留当前节点
        return head.next if head.val == val else head
```

### 逐行解释：
1. **递归的基本条件**：
   - **`if not head:`**：如果链表为空，返回 `None`。这是递归的结束条件，当链表被完全遍历到尾部时，停止递归。

2. **递归处理链表的下一节点**：
   - **`head.next = self.removeElements(head.next, val)`**：递归调用处理链表的下一个节点，通过递归将链表逐层剥离，直到处理到最后一个节点。

3. **删除当前节点（如果需要）**：
   - **`return head.next if head.val == val else head`**：如果当前节点的值等于要删除的目标值 `val`，则返回当前节点的下一节点 `head.next`，跳过当前节点。
   - 否则返回当前节点 `head`，保留该节点。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(n)`，其中 `n` 是链表的长度。我们需要遍历每个节点一次，因此时间复杂度是线性。
- **空间复杂度**：`O(n)`，由于递归的深度与链表的长度成正比，每次递归都会占用栈空间，空间复杂度为 `O(n)`。

### 关键概念：
- **递归删除**：通过递归处理链表，先处理后面的节点，再根据当前节点的值决定是否跳过或保留该节点。
- **链表操作**：递归方法在处理链表删除时更加简洁，但会增加递归栈的使用。

### 小结：
- 递归方法通过对链表的逐层处理，实现了对每个节点的检查和删除，代码简洁且直观。

### Tips:
- 递归可以简化链表的处理，但需要注意递归深度，避免在深度很大的链表中导致栈溢出。

### Warning:
- 递归的基准条件是链表为空时，务必注意处理好链表的边界条件。
