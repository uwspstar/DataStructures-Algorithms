509. Fibonacci Number

https://leetcode.com/problems/fibonacci-number/description/

The Fibonacci numbers, commonly denoted F(n) form a sequence, called the Fibonacci sequence, such that each number is the sum of the two preceding ones, starting from 0 and 1. That is,

F(0) = 0, F(1) = 1
F(n) = F(n - 1) + F(n - 2), for n > 1.
Given n, calculate F(n).

 

Example 1:

Input: n = 2
Output: 1
Explanation: F(2) = F(1) + F(0) = 1 + 0 = 1.
Example 2:

Input: n = 3
Output: 2
Explanation: F(3) = F(2) + F(1) = 1 + 1 = 2.
Example 3:

Input: n = 4
Output: 3
Explanation: F(4) = F(3) + F(2) = 2 + 1 = 3.

````python
class Solution:
    def fib(self, n:int) -> int:
        if n == 0:
            return 0
        if n == 1:
            return 1

        v1, v2 = 1, 0
        for i in range(2, n + 1):
            tmp = v1
            v1 = v1 + v2
            v2 = tmp

        return v1

# T(n)
# S(1)
````

```python
class Solution:
    def fib(self, n: int) -> int:
        # 如果 n 为 0，直接返回 0，因为斐波那契数列的第 0 项是 0
        if n == 0:
            return 0
        
        # 如果 n 为 1，直接返回 1，因为斐波那契数列的第 1 项是 1
        if n == 1:
            return 1

        # 初始化 v1 和 v2，分别表示当前和前一个斐波那契数
        v1, v2 = 1, 0
        
        # 从第 2 项开始，计算斐波那契数
        for i in range(2, n + 1):
            # 临时存储当前 v1 的值，作为下一轮循环中的 v2
            tmp = v1
            # 当前的斐波那契数等于前两个数之和
            v1 = v1 + v2
            # 更新 v2 为当前的 v1
            v2 = tmp

        # 最终返回 v1，即第 n 项斐波那契数
        return v1
```

### 逐行解释：

1. **特殊情况**：
   - `if n == 0` 和 `if n == 1`：直接处理斐波那契数列的前两项，返回 `0` 和 `1`，这是斐波那契数列的定义部分。

2. **初始化**：
   - `v1` 和 `v2` 分别表示当前和前一个斐波那契数，初始值为 `1` 和 `0`。这两个值用于计算后续的斐波那契数。

3. **迭代计算斐波那契数列**：
   - `for i in range(2, n + 1)`：从第 2 项开始，通过迭代计算每一项的斐波那契数。
   - `tmp = v1`：临时存储 `v1` 的值，以便下一轮更新 `v2`。
   - `v1 = v1 + v2`：更新 `v1` 为当前斐波那契数，即前两个数之和。
   - `v2 = tmp`：将 `v2` 更新为上一轮的 `v1`。

4. **返回结果**：
   - 最终返回 `v1`，即第 `n` 项斐波那契数。

### 时间和空间复杂度：

- **时间复杂度**：`O(n)`，因为我们通过迭代计算斐波那契数列，每次计算的时间是常数，因此整体的时间复杂度是线性的。
  
- **空间复杂度**：`O(1)`，只使用了常数级别的变量 `v1`、`v2` 和 `tmp`，没有额外的空间开销，因此空间复杂度为常数。

### 结论：
这个算法通过迭代计算斐波那契数列，避免了递归的高空间和时间复杂度问题。通过维护两个变量来记录前两项的值，能够以线性的时间和常数的空间复杂度高效计算第 `n` 项的斐波那契数。

````python
class Solution:
    def __init__(self):
        self.memo = {}

    def fib(self, n:int) -> int:
        if n == 0:
            return 0
        if n == 1:
            return 1
        
        if n not in self.memo: 
            self.memo[n] = self.fib(n - 1) + self.fib(n - 2)

        return self.memo[n]


# T(n)
# S(n)
````

2. **递归基准条件**：
   - `if n == 0` 和 `if n == 1`：处理斐波那契数列的前两项，直接返回 `0` 和 `1`，这是递归的基准条件。

3. **记忆化递归**：

4. **返回结果**：

### 时间和空间复杂度：

- **时间复杂度**：`O(n)`。每个 `n` 只被计算一次，计算后存储在 `memo` 中，因此时间复杂度是线性的。
