234. Palindrome Linked List

https://leetcode.com/problems/palindrome-linked-list/description/

Given the head of a singly linked list, return true if it is a palindrome or false otherwise.

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        if not head or not head.next:
            return True
        
        slow, fast = head, head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        prev = None
        while slow:
            nxt = slow.next
            slow.next = prev
            prev = slow
            slow = nxt
        
        left, right = head, prev
        while right:
            if left.val != right.val:
                return False
            left = left.next
            right = right.next
        
        return True
````

以下是代码的逐行解释，包含中文注释：

```python
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        # 如果链表为空或者只有一个节点，直接返回 True，因为它已经是回文
        if not head or not head.next:
            return True
        
        # 初始化两个指针，慢指针 slow 和快指针 fast
        slow, fast = head, head
        # 快指针每次移动两步，慢指针移动一步，目的是找到链表的中间位置
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        # 反转链表的后半部分，初始化 prev 为 None
        prev = None
        while slow:
            nxt = slow.next  # 暂存下一个节点
            slow.next = prev  # 将当前节点指向前一个节点，逐步反转链表
            prev = slow  # prev 移动到当前节点
            slow = nxt  # 继续向前遍历链表
        
        # 比较链表的前半部分和反转后的后半部分
        left, right = head, prev
        while right:
            # 如果左右两部分的值不同，返回 False，表示不是回文
            if left.val != right.val:
                return False
            left = left.next  # 前半部分继续向前
            right = right.next  # 反转后的后半部分也继续向前
        
        # 如果所有节点都匹配，返回 True，表示链表是回文
        return True
```

### 逐行解释：

1. **特殊情况处理**：
   - 如果链表为空或者只有一个节点，那么它自然是回文的，因此直接返回 `True`。

2. **寻找链表中点**：
   - 使用两个指针 `slow` 和 `fast`，`slow` 每次移动一步，`fast` 每次移动两步。通过这种方式，`slow` 在 `fast` 到达链表末尾时，正好位于链表的中间。

3. **反转链表的后半部分**：
   - 从链表的中间位置开始，逐步将后半部分的节点指向前一个节点，最终 `prev` 会指向反转后的链表的头。

4. **比较前半部分和反转后的后半部分**：
   - 使用两个指针 `left` 和 `right` 分别指向链表的头和反转后的链表头。依次比较对应节点的值，如果有不相等的情况，返回 `False`。

5. **返回结果**：
   - 如果所有节点都匹配，则返回 `True`，表示链表是回文。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(n)`，其中 `n` 是链表中的节点数。
  - 我们遍历链表两次：第一次是找到链表的中点，第二次是比较链表的前后半部分。
  
- **空间复杂度**：`O(1)`，我们只使用了常量级的额外空间。

因此，整体的时间复杂度是 `O(n)`，空间复杂度是 `O(1)`。

This example demonstrates how to check if a linked list is a palindrome by using two-pointer techniques and reversing part of the list.

````python
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        if not head or not head.next:
            return True

        # Step 1: Find the middle of the list
        middle = self.findMiddle(head)
        
        # Step 2: Reverse the second half of the list
        reversed_head = self.reverseList(middle)
        
        # Step 3: Compare the first half and the reversed second half
        return self.compareLists(head, reversed_head)

    def findMiddle(self, head: ListNode) -> ListNode:
        slow, fast = head, head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow

    def reverseList(self, head: ListNode) -> ListNode:
        prev = None
        while head:
            nxt = head.next
            head.next = prev
            prev = head
            head = nxt
        return prev

    def compareLists(self, head1: ListNode, head2: ListNode) -> bool:
        while head2:  # Only compare the second half
            if head1.val != head2.val:
                return False
            head1 = head1.next
            head2 = head2.next
        return True
````
