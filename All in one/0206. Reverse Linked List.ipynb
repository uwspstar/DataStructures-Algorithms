206. Reverse Linked List

https://leetcode.com/problems/reverse-linked-list/description/

Given the head of a singly linked list, reverse the list, and return the reversed list.

 

Example 1:


Input: head = [1,2,3,4,5]
Output: [5,4,3,2,1]
Example 2:


Input: head = [1,2]
Output: [2,1]
Example 3:

Input: head = []
Output: []

````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:

        pre, cur = None, head
        while cur:
            nxt = cur.next
            cur.next = pre
            pre = cur
            cur = nxt

        return pre
````

以下是代码的逐行解释，包含中文注释：

```python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # 初始化 pre 为 None，cur 为链表的头节点 head
        pre, cur = None, head
        
        # 遍历整个链表，直到 cur 为空
        while cur:
            # 保存当前节点的下一个节点
            nxt = cur.next
            # 将当前节点的 next 指针指向前一个节点，实现链表反转
            cur.next = pre
            # 更新 pre 为当前节点，作为下一次迭代中当前节点的前一节点
            pre = cur
            # 移动 cur 到下一个节点，继续迭代
            cur = nxt

        # 当 cur 为空时，pre 就是反转后的链表头节点，返回 pre
        return pre
```

### 逐行解释：

1. **初始化指针**：
   - `pre, cur = None, head`：`pre` 用于指向反转链表的前一节点，初始为 `None`，因为反转后头节点的 `next` 应该指向 `None`。`cur` 指向链表的当前节点，初始为链表头节点 `head`。

2. **主循环**：
   - `while cur:`：当 `cur` 不为空时，遍历链表的每个节点，进行反转。

3. **保存下一个节点**：
   - `nxt = cur.next`：在反转当前节点指针之前，先保存当前节点的下一个节点 `nxt`，这样不会丢失原始链表的顺序。

4. **反转指针**：
   - `cur.next = pre`：将当前节点的 `next` 指向 `pre`，实现指针反转。

5. **移动指针**：
   - `pre = cur`：将 `pre` 更新为当前节点 `cur`，表示当前节点已经完成反转，并将在下一次迭代中作为前一节点。
   - `cur = nxt`：将 `cur` 移动到原链表中的下一个节点 `nxt`，继续循环。

6. **返回反转后的链表头节点**：
   - `return pre`：当 `cur` 为 `None` 时，链表已经全部反转，`pre` 此时指向反转后的链表头节点，因此返回 `pre`。

### 时间和空间复杂度：

- **时间复杂度**：`O(n)`，其中 `n` 是链表的长度。我们遍历链表一次，每个节点的操作是常数时间，因此时间复杂度是线性的。
  
- **空间复杂度**：`O(1)`，因为该算法是原地反转链表，只使用了几个额外的指针变量（`pre`, `cur`, `nxt`），因此空间复杂度是常数。

### 结论：
该算法通过迭代的方式反转链表，使用两个指针 `pre` 和 `cur` 来逐步改变每个节点的 `next` 指针方向。最终返回反转后的链表头节点。该算法时间复杂度为 `O(n)`，空间复杂度为 `O(1)`，是一种高效的链表反转方法。

````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head

        new_head = self.reverseList(head.next)    
        head.next.next = head
        head.next = None

        return new_head
````

以下是代码的逐行解释，包含中文注释：

```python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # 如果链表为空，或者只有一个节点，直接返回 head（无需反转）
        if not head or not head.next:
            return head

        # 递归反转子链表，直到找到最后一个节点（新的头节点 new_head）
        new_head = self.reverseList(head.next)
        
        # 反转当前节点与下一个节点的指针
        head.next.next = head  # 将 head.next 指向 head
        head.next = None       # 断开原本的 head 与 head.next 的连接

        # 返回新的头节点
        return new_head
```

### 逐行解释：

1. **递归终止条件**：
   - `if not head or not head.next`：如果 `head` 为空，或者 `head` 是链表的最后一个节点，直接返回 `head`。这是递归的基准条件，意味着我们已经找到了链表的末尾，开始反转。

2. **递归反转子链表**：
   - `new_head = self.reverseList(head.next)`：递归调用 `reverseList`，继续反转从 `head.next` 开始的链表。当递归回溯时，`new_head` 会是反转后的链表的头节点。

3. **反转当前节点指针**：
   - `head.next.next = head`：将 `head.next` 的 `next` 指针指向 `head`，这一步实现了指针反转，即将 `head` 连接到它的下一个节点后面。
   - `head.next = None`：将 `head` 的 `next` 设置为 `None`，因为 `head` 将成为反转后链表的末尾节点，它的 `next` 应该指向 `None`。

4. **返回新的头节点**：
   - `return new_head`：返回反转后的链表头节点。由于每次递归都返回最终的头节点，因此每次回溯时都会返回 `new_head`。

### 时间和空间复杂度：

- **时间复杂度**：`O(n)`，其中 `n` 是链表的长度。每个节点被访问一次，因此时间复杂度为线性。
  
- **空间复杂度**：`O(n)`，由于使用了递归，每次递归调用都需要存储函数调用栈，最坏情况下递归深度为链表的长度，因此空间复杂度为 `O(n)`。

### 与迭代版本的对比：
- **递归版本**：递归版本通过递归调用函数，利用系统的调用栈逐步反转链表。虽然逻辑清晰，代码简洁，但由于递归的深度受限于链表的长度，可能会导致栈溢出问题，尤其是在处理非常长的链表时。
- **迭代版本**：迭代版本使用常数空间，通过循环改变指针方向，空间复杂度为 `O(1)`。相比递归版本，迭代版本更节省内存。

### 结论：
该算法通过递归反转链表，每次反转当前节点与下一个节点的指针，最终返回新的头节点。时间复杂度为 `O(n)`，空间复杂度为 `O(n)`。如果链表非常长，迭代版本可能更适合。
