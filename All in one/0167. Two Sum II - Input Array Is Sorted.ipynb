167. Two Sum II - Input Array Is Sorted

https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/description/

Given a 1-indexed array of integers numbers that is already sorted in non-decreasing order, find two numbers such that they add up to a specific target number. Let these two numbers be numbers[index1] and numbers[index2] where 1 <= index1 < index2 <= numbers.length.

Return the indices of the two numbers, index1 and index2, added by one as an integer array [index1, index2] of length 2.

The tests are generated such that there is exactly one solution. You may not use the same element twice.

Your solution must use only constant extra space.

 

Example 1:

Input: numbers = [2,7,11,15], target = 9
Output: [1,2]
Explanation: The sum of 2 and 7 is 9. Therefore, index1 = 1, index2 = 2. We return [1, 2].
Example 2:

Input: numbers = [2,3,4], target = 6
Output: [1,3]
Explanation: The sum of 2 and 4 is 6. Therefore index1 = 1, index2 = 3. We return [1, 3].
Example 3:

Input: numbers = [-1,0], target = -1
Output: [1,2]
Explanation: The sum of -1 and 0 is -1. Therefore index1 = 1, index2 = 2. We return [1, 2].

````python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        l, r = 0, len(numbers) - 1
        while l < r:
            curr = numbers[l] + numbers[r]
            if curr == target:
                return [l + 1, r + 1]
            elif curr < target:
                l += 1
            else:
                r -= 1
        return []
````

以下是代码的逐行解释，包含中文注释：

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        # 初始化两个指针 l 和 r，分别指向数组的起始位置和末尾位置
        l, r = 0, len(numbers) - 1
        
        # 当 l 小于 r 时继续循环
        while l < r:
            # 计算当前两个指针对应元素的和
            curr = numbers[l] + numbers[r]
            
            # 如果当前和等于目标值，返回两个指针的位置（1-based 索引）
            if curr == target:
                return [l + 1, r + 1]
            
            # 如果当前和小于目标值，说明需要更大的数，所以移动左指针 l
            elif curr < target:
                l += 1
            
            # 如果当前和大于目标值，说明需要更小的数，所以移动右指针 r
            else:
                r -= 1
        
        # 如果没有找到符合条件的两个数，返回空列表
        return []
```

### 逐行解释：

1. **初始化指针**：
   - `l, r = 0, len(numbers) - 1`：`l` 指向数组的起始位置（最小值），`r` 指向数组的末尾位置（最大值）。通过使用双指针技术，我们可以在有序数组中高效地查找两个数的和是否等于目标值。

2. **主循环**：
   - `while l < r`：只要左指针小于右指针，继续搜索。

3. **计算当前和**：
   - `curr = numbers[l] + numbers[r]`：计算 `numbers[l]` 和 `numbers[r]` 的和，判断其是否等于目标值。

4. **判断当前和**：
   - `if curr == target`：如果当前和等于目标值，返回两个指针的 1-based 索引位置（题目要求返回从 1 开始的索引，因此返回 `[l + 1, r + 1]`）。
   
   - `elif curr < target`：如果当前和小于目标值，说明需要一个更大的数，所以移动左指针 `l` 向右，`l += 1`。
   
   - `else`：如果当前和大于目标值，说明需要一个更小的数，所以移动右指针 `r` 向左，`r -= 1`。

5. **返回结果**：
   - 如果最终没有找到符合条件的两个数，返回空列表 `[]`。不过在该问题中，题目假设一定有解，因此实际上不会执行到这一行。

### 时间和空间复杂度：

- **时间复杂度**：`O(n)`，其中 `n` 是 `numbers` 列表的长度。我们只需要一次遍历，并且每次操作都会移动一个指针，因此时间复杂度是线性的。
  
- **空间复杂度**：`O(1)`，除了输入和输出之外，算法只使用了常数级别的额外空间用于存储指针和临时变量。

### 结论：
该算法通过双指针技术高效地解决了“两数之和”问题，利用数组已经排序的特性。通过调整左、右指针，能够快速找到两个数的和等于目标值的解，时间复杂度为 `O(n)`，且空间复杂度为 `O(1)`。
