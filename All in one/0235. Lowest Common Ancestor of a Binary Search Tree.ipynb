235. Lowest Common Ancestor of a Binary Search Tree

https://leetcode.com/problems/lowest-common-ancestor-of-a-binary-search-tree/description/

Given a binary search tree (BST), find the lowest common ancestor (LCA) node of two given nodes in the BST.

According to the definition of LCA on Wikipedia: “The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself).”

 

Example 1:


Input: root = [6,2,8,0,4,7,9,null,null,3,5], p = 2, q = 8
Output: 6
Explanation: The LCA of nodes 2 and 8 is 6.
Example 2:


Input: root = [6,2,8,0,4,7,9,null,null,3,5], p = 2, q = 4
Output: 2
Explanation: The LCA of nodes 2 and 4 is 2, since a node can be a descendant of itself according to the LCA definition.
Example 3:

Input: root = [2,1], p = 2, q = 1
Output: 2
 

Constraints:

The number of nodes in the tree is in the range [2, 105].
-109 <= Node.val <= 109
All Node.val are unique.
p != q
p and q will exist in the BST.

````python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        cur = root
        while cur:
            if p.val > cur.val and q.val > cur.val:
                cur = cur.right
            elif p.val < cur.val and q.val < cur.val:
                cur = cur.left
            else:
                return cur
            
````

```python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        # 初始化当前节点为根节点
        cur = root

        # 开始遍历树，寻找最小公共祖先
        while cur:
            # 如果 p 和 q 的值都大于当前节点的值，说明它们在右子树
            if p.val > cur.val and q.val > cur.val:
                cur = cur.right  # 移动到右子树
            # 如果 p 和 q 的值都小于当前节点的值，说明它们在左子树
            elif p.val < cur.val and q.val < cur.val:
                cur = cur.left  # 移动到左子树
            # 否则，当前节点就是它们的最小公共祖先
            else:
                return cur  # 返回当前节点
```

### 逐行解释：
1. **初始化**：
   - `cur = root`: 将当前节点初始化为树的根节点 `root`。
   - 这是用于遍历树以寻找 `p` 和 `q` 的最小公共祖先（LCA）。

2. **遍历树**：
   - 使用 `while cur` 循环，遍历树直到找到 LCA 或者到达叶节点。
   - **`if p.val > cur.val and q.val > cur.val`**：
     - 如果 `p` 和 `q` 的值都大于当前节点 `cur` 的值，说明 `p` 和 `q` 在当前节点的右子树中，因此我们移动到 `cur.right`。
   - **`elif p.val < cur.val and q.val < cur.val`**：
     - 如果 `p` 和 `q` 的值都小于当前节点 `cur` 的值，说明 `p` 和 `q` 在当前节点的左子树中，因此我们移动到 `cur.left`。
   - **`else`**：
     - 如果 `p` 和 `q` 分别位于当前节点的两侧，或者其中一个就是当前节点，则当前节点就是它们的最小公共祖先，直接返回当前节点。

3. **返回 LCA**：
   - 一旦找到满足条件的节点，返回该节点作为 `p` 和 `q` 的最小公共祖先。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(h)`，其中 `h` 是树的高度。在最坏情况下，树的高度可以是 `O(n)`（线性树），但在平衡二叉树中是 `O(log n)`。
- **空间复杂度**：`O(1)`，我们只使用了常量级的额外空间。

### 关键概念：
- **二叉搜索树的性质**：在二叉搜索树（BST）中，左子树的所有节点值都小于根节点，右子树的所有节点值都大于根节点。
- **最小公共祖先（LCA）**：如果节点 `p` 和 `q` 分别位于当前节点的两侧，那么当前节点就是它们的 LCA。

### 小结：
- 通过根据二叉搜索树的性质，我们可以通过不断移动左右子树来找到两个节点的最小公共祖先。

### Tips:
- 如果当前节点的值位于 `p` 和 `q` 之间或等于其中之一，那么它就是 LCA。

### Warning:
- 在树的遍历过程中要注意避免无限循环，当 `p` 或 `q` 之一就是根节点时，直接返回根节点作为 LCA。

````python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if not root:
            return root
        
        if root.val < p.val and root.val < q.val:
            return self.lowestCommonAncestor(root.right, p, q)
        if root.val > p.val and root.val > q.val:
            return self.lowestCommonAncestor(root.left, p, q)

        return root
````
