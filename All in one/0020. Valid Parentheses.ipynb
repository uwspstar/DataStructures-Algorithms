20. Valid Parentheses

https://leetcode.com/problems/valid-parentheses/description/


Given a string s containing just the characters (, ), {, }, [ and ], determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Example:

Input: s = "()"
Output: true
Example:

Input: s = "()[]{}"
Output: true
Example:

Input: s = "(]"
Output: false

In Python, map is not a keyword but a built-in function.

````python
class Solution: # 1
    def isValid(self, s: str) -> bool:
        stack = []
        mapping = {')':'(', ']': '[', '}':'{'} # 初始化一个空栈来跟踪开括号。同时，我们创建一个映射，将闭括号与其对应的开括号关联。
        for c in s:
            if c in mapping: # 通过在 mapping 中查找当前字符来检查它是否为闭括号。
                top = stack.pop() if stack else '#'
                if mapping[c] != top:
                    return False
            else:
                stack.append(c)
        
        return not stack
````

````python
class Solution: # 2 mapping is different
    def isValid(self, s:str) -> bool:
        stack = []
        pairs ={"(":")", "[": "]", "{":"}"}

        for c in s:
            if c in pairs:
                stack.append(c)
            else:
                if len(stack) == 0 or pairs[stack.pop()] != c:
                    return False
        
        return len(stack) == 0
        



````

以下是代码的逐行解释，包含中文注释：

```python
class Solution: # 这里的 2 mapping 不同：用 pairs 字典进行括号配对
    def isValid(self, s: str) -> bool:
        # 创建一个空的栈，用来存储左括号
        stack = []
        
        # 定义一个字典 pairs，用来存储每种左括号及其对应的右括号
        pairs = {"(": ")", "[": "]", "{": "}"}

        # 遍历字符串 s 中的每一个字符
        for c in s:
            # 如果字符是左括号，将其放入栈中
            if c in pairs:
                stack.append(c)
            else:
                # 如果字符是右括号，检查栈是否为空或栈顶元素是否匹配当前右括号
                if len(stack) == 0 or pairs[stack.pop()] != c:
                    # 如果栈为空或配对不正确，返回 False
                    return False
        
        # 遍历结束后，检查栈是否为空，只有在所有括号正确闭合时栈才为空
        return len(stack) == 0
```

### 逐行解释：
1. **初始化**：
   - 使用一个空列表 `stack` 来模拟栈操作，用于保存遇到的左括号。
   - 定义一个字典 `pairs`，将左括号作为键，右括号作为值，用来快速查找每个左括号对应的右括号。

2. **遍历字符串 `s`**：
   - 对于字符串 `s` 中的每个字符 `c`：
     - 如果 `c` 是左括号（存在于 `pairs` 的键中），将其压入栈 `stack`。
     - 如果 `c` 是右括号，检查栈的状态：
       - 如果栈为空，说明此时有右括号但没有左括号与之匹配，返回 `False`。
       - 如果栈不为空，弹出栈顶元素，检查它与当前的右括号是否匹配。如果不匹配，返回 `False`。

3. **最终检查栈是否为空**：
   - 遍历完所有字符后，如果栈为空，说明所有括号都匹配，返回 `True`。
   - 如果栈不为空，说明还有未匹配的左括号，返回 `False`。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(n)`，其中 `n` 是字符串 `s` 的长度。
  - 我们只遍历字符串一次，对于每个字符，执行入栈、出栈和字典查找等操作，这些操作的时间复杂度都是 `O(1)`。
  
- **空间复杂度**：`O(n)`，最坏情况下，栈中会存储所有的左括号，因此空间复杂度是线性的。

因此，整体的时间复杂度是 `O(n)`，空间复杂度也是 `O(n)`。
