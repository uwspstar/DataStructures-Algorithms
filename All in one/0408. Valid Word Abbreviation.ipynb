408. Valid Word Abbreviation

https://leetcode.com/problems/valid-word-abbreviation/description/

A string can be abbreviated by replacing any number of non-adjacent, non-empty substrings with their lengths. The lengths should not have leading zeros.

For example, a string such as "substitution" could be abbreviated as (but not limited to):

"s10n" ("s ubstitutio n")
"sub4u4" ("sub stit u tion")
"12" ("substitution")
"su3i1u2on" ("su bst i t u ti on")
"substitution" (no substrings replaced)
The following are not valid abbreviations:

"s55n" ("s ubsti tutio n", the replaced substrings are adjacent)
"s010n" (has leading zeros)
"s0ubstitution" (replaces an empty substring)
Given a string word and an abbreviation abbr, return whether the string matches the given abbreviation.

A substring is a contiguous non-empty sequence of characters within a string.

 

Example 1:

Input: word = "internationalization", abbr = "i12iz4n"
Output: true
Explanation: The word "internationalization" can be abbreviated as "i12iz4n" ("i nternational iz atio n").
Example 2:

Input: word = "apple", abbr = "a2e"
Output: false
Explanation: The word "apple" cannot be abbreviated as "a2e".


````python
class Solution:
    def validWordAbbreviation(self, word: str, abbr: str) -> bool:
        i, j = 0, 0
        
        while i < len(word) and j < len(abbr):
            if abbr[j].isalpha(): # letter
                if word[i]!= abbr[j]:
                    return False
                i += 1
                j += 1
            else:
                if abbr[j] =='0':
                    return False
                
                number = 0
                
                while j < len(abbr) and abbr[j].isdigit():
                    number = number * 10 + int(abbr[j])
                    j += 1
                
                i += number
                
        return i == len(word) and j == len(abbr)
````

````python
class Solution:
    def validWordAbbreviation(self, word, abbr):
        i = j = 0
        m, n = len(word), len(abbr)
        while i < m and j < n:
            if word[i] == abbr[j]:
                i += 1
                j += 1
            elif abbr[j] == "0":
                return False
            elif abbr[j].isnumeric():
                k = j
                while k < n and abbr[k].isnumeric():
                    k += 1
                i += int(abbr[j:k])
                j = k
            else:
                return False
        return i == m and j == n
    
# https://leetcode.com/problems/valid-word-abbreviation/solutions/144195/python-straightforward-solution-beats-100/
````

以下是代码的逐行解释，包含中文注释：

```python
class Solution:
    def validWordAbbreviation(self, word, abbr):
        # 初始化两个指针 i 和 j，分别用于遍历 word 和 abbr
        i = j = 0
        # 获取 word 和 abbr 的长度
        m, n = len(word), len(abbr)
        
        # 当两个指针都在字符串范围内时继续循环
        while i < m and j < n:
            # 如果 word 和 abbr 的当前字符相等，移动两个指针
            if word[i] == abbr[j]:
                i += 1
                j += 1
            # 如果 abbr 以 '0' 开头，则无效，因为缩写不能以 '0' 开头
            elif abbr[j] == "0":
                return False
            # 如果 abbr 的当前字符是数字，处理缩写
            elif abbr[j].isnumeric():
                # 使用指针 k 来查找数字的结束位置
                k = j
                # 找到 abbr 中数字的范围，k 最终指向第一个不是数字的字符
                while k < n and abbr[k].isnumeric():
                    k += 1
                # 将 i 指针向后移动当前数字的大小，表示跳过相应的字符
                i += int(abbr[j:k])
                # 将 j 指针移动到数字后的下一个字符
                j = k
            # 如果字符不匹配且 abbr 当前字符也不是数字，返回 False
            else:
                return False
        
        # 如果 i 和 j 都指向了 word 和 abbr 的末尾，说明完全匹配，返回 True
        return i == m and j == n
```

### 逐行解释：

1. **初始化**：
   - `i` 和 `j` 分别初始化为 `0`，用于遍历 `word` 和 `abbr`。
   - `m` 和 `n` 分别是 `word` 和 `abbr` 的长度，用来控制指针的范围。

2. **主循环**：
   - `while i < m and j < n`：只要 `i` 和 `j` 还在字符串的范围内，继续比较 `word` 和 `abbr`。
   
3. **字符匹配**：
   - `if word[i] == abbr[j]`：如果 `word[i]` 和 `abbr[j]` 当前字符相同，则分别移动 `i` 和 `j`，继续比较下一个字符。

4. **非法的前导 0**：
   - `elif abbr[j] == "0"`：如果 `abbr` 中数字以 `0` 开头，这表示无效，因为数字不能以 `0` 开头，直接返回 `False`。

5. **处理数字缩写**：
   - `elif abbr[j].isnumeric()`：如果 `abbr[j]` 是数字，开始处理数字部分。
   - `while k < n and abbr[k].isnumeric()`：使用 `k` 指针来找到数字的结束位置，直到 `abbr[k]` 不是数字为止。
   - `i += int(abbr[j:k])`：将 `i` 向后移动数字的大小，跳过 `word` 中相应的字符。

6. **处理字符不匹配**：
   - `else`：如果 `abbr[j]` 既不是字母匹配，也不是数字缩写，则说明无法匹配，返回 `False`。

7. **最终检查**：
   - `return i == m and j == n`：在循环结束后，检查 `i` 是否已经遍历完 `word`，`j` 是否已经遍历完 `abbr`。如果两者都到达末尾，则说明缩写完全匹配，返回 `True`；否则返回 `False`。

### 时间和空间复杂度分析：

- **时间复杂度**：`O(n + m)`，其中 `n` 是 `abbr` 的长度，`m` 是 `word` 的长度。我们遍历 `word` 和 `abbr` 的每个字符，时间复杂度是线性的。
  
- **空间复杂度**：`O(1)`，除了输入字符串，我们只使用了常数空间来存储指针 `i`、`j` 和 `k` 以及临时变量，因此空间复杂度是常数。

### 结论：
这个算法通过双指针的方法遍历 `word` 和 `abbr`，逐步比较字符或处理缩写数字，最终返回是否有效的缩写匹配。
