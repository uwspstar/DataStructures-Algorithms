202. Happy Number

https://leetcode.com/problems/happy-number/description/


Write an algorithm to determine if a number n is happy.

A happy number is a number defined by the following process:

Starting with any positive integer, replace the number by the sum of the squares of its digits.
Repeat the process until the number equals 1 (where it will stay), or it loops endlessly in a cycle which does not include 1.
Those numbers for which this process ends in 1 are happy.
Return true if n is a happy number, and false if not.

 

Example 1:

Input: n = 19
Output: true
Explanation:
12 + 92 = 82
82 + 22 = 68
62 + 82 = 100
12 + 02 + 02 = 1
Example 2:

Input: n = 2
Output: false
 

Constraints:

1 <= n <= 231 - 1

````python
class Solution:
    def isHappy(self, n: int) -> bool:
        if n == 1:
            return True

        see = set()
        
        while n not in see:
            see.add(n)
            n = self.sum_of_squares(n) 
            if n == 1:
                return True
        return False

    def sum_of_squares(self, n:int):
        res = 0
        while n:
            digit = n % 10
            digit = digit**2
            res += digit
            n = n // 10
        return res

        
````

```python
class Solution:
    def isHappy(self, n: int) -> bool:
        # 如果 n 等于 1，直接返回 True
        if n == 1:
            return True

        # 使用一个集合来存储我们已经见过的数字，防止进入无限循环
        see = set()
        
        # 当 n 不在集合中时，继续循环
        while n not in see:
            see.add(n)  # 将当前数字添加到集合中
            n = self.sum_of_squares(n)  # 计算每个数字平方和，更新 n
            # 如果 n 等于 1，说明是快乐数，返回 True
            if n == 1:
                return True

        # 如果进入循环，最终没有得到 1，则返回 False
        return False

    # 计算数字 n 的每一位的平方和
    def sum_of_squares(self, n: int):
        res = 0  # 用于存储平方和的结果
        while n:
            digit = n % 10  # 提取最后一位数字
            digit = digit ** 2  # 计算该位数字的平方
            res += digit  # 将平方值加入结果
            n = n // 10  # 移除该位数字
        return res  # 返回所有位平方和的结果
```

### 逐行解释：
1. **`isHappy` 方法**：
   - **`if n == 1`**：如果 `n` 是 1，说明是快乐数，直接返回 `True`。
   - **`see = set()`**：创建一个集合 `see` 来存储已经计算过的数字，防止陷入无限循环。

2. **主循环**：
   - **`while n not in see`**：当 `n` 还没有被处理时，继续循环。
   - **`see.add(n)`**：将当前数字 `n` 加入集合，防止后续重复处理。
   - **`n = self.sum_of_squares(n)`**：调用 `sum_of_squares` 方法，计算当前数字的每一位的平方和，更新 `n`。
   - **`if n == 1`**：如果在计算过程中 `n` 变成 1，说明该数是快乐数，返回 `True`。
   - 如果最终没有找到 1，说明不是快乐数，返回 `False`。

3. **`sum_of_squares` 方法**：
   - **`res = 0`**：初始化结果为 0。
   - **`digit = n % 10`**：提取 `n` 的最后一位数字。
   - **`digit ** 2`**：计算该位数字的平方并累加到结果 `res` 中。
   - **`n = n // 10`**：移除已经处理的最后一位数字，继续处理剩下的数字。
   - 最终返回所有位数字的平方和。

### 时间复杂度 (Big O) 分析：
- **时间复杂度**：`O(log n)`，因为每次我们处理 `n` 的每一位数字并求其平方和，直到 `n` 变成 1 或进入循环。
- **空间复杂度**：`O(log n)`，集合 `see` 存储已经处理过的数字。

### 关键概念：
- **快乐数**：通过对一个数字的各位数字求平方和并不断重复该过程，如果最终得到 `1`，那么这个数字就是快乐数，否则它将陷入循环。
- **使用集合防止循环**：通过使用集合 `see`，我们能够检测到是否进入无限循环，从而避免死循环。

### 小结：
- 通过求平方和并利用集合存储计算过的数字，我们可以有效判断一个数字是否为快乐数。

### Tips:
- 使用集合存储已经处理过的数字是避免无限循环的常用技巧。
